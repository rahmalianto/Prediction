```python
import csv
import numpy as np
import pandas as pd
from chefboost import Chefboost as chef
```

```python
data = pd.read_csv('bank-additional/bank-additional.csv', sep=';')
# data.head()
training_set = data.loc[0:999]
test_set = data.loc[999:4119]
training_set.head()
```

```python
# Create Model and training
config = {'algorithm': 'ID3'}
model = chef.fit(training_set, config) 
```

```python
# Testing model
correct = 0
wrong = 0
for index, row in test_set.iterrows():
    prediction = chef.predict(model, row)
    target = row['Decision']
    if prediction == target:
        correct += 1
    else:
        wrong += 1
        
print("accuracy: %f. Correct: %d. Wrong: %d" %(float(correct/(correct+wrong)), correct, wrong))
```
